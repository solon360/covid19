```python
#Importação das bibliotecas
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
#Recebimento das tabelas
brasil = pd.read_csv('https://raw.githubusercontent.com/solon360/covid19/main/brasil2.csv')
mexico = pd.read_csv('https://raw.githubusercontent.com/solon360/covid19/main/mexico2.csv')
novazelandia = pd.read_csv('https://raw.githubusercontent.com/solon360/covid19/main/novazelandia2.csv')
restricoes = pd.read_csv('https://raw.githubusercontent.com/solon360/covid19/main/restricao.csv')

#Remoção das colunas que estão inteiramente vazias
brasil.dropna(how='all', axis=1, inplace = True)
mexico.dropna(how='all', axis=1, inplace = True)
novazelandia.dropna(how='all', axis=1, inplace = True)
#brasil.info()
```

```python
#Novos casos diários
novos_casos_mexico = px.line(mexico, x="date", y="new_cases_smoothed", color='location')
novos_casos_brasil = px.line(brasil, x="date", y="new_cases_smoothed", color='location')
novos_casos_novazelandia = px.line(novazelandia, x="date", y="new_cases_smoothed", color='location')
novos_casos = go.Figure(data = novos_casos_mexico.data + novos_casos_brasil.data + novos_casos_novazelandia.data)
novos_casos.data[0].line.color = 'rgb(230, 5, 42)'
novos_casos.data[1].line.color = 'rgb(162, 227, 11)'
novos_casos.data[2].line.color = 'rgb(15, 32, 214)'
novos_casos.show()
```

```python
#Novas mortes diárias
novas_mortes_mexico = px.line(mexico, x="date", y="new_deaths_smoothed", color='location')
novas_mortes_brasil = px.line(brasil, x="date", y="new_deaths_smoothed", color='location')
novas_mortes_novazelandia = px.line(novazelandia, x="date", y="new_deaths_smoothed", color='location')
novas_mortes = go.Figure(data = novas_mortes_mexico.data + novas_mortes_brasil.data + novas_mortes_novazelandia.data)
novas_mortes.data[0].line.color = 'rgb(230, 5, 42)'
novas_mortes.data[1].line.color = 'rgb(162, 227, 11)'
novas_mortes.data[2].line.color = 'rgb(15, 32, 214)'
novas_mortes.show()
```

```python
#Índice de rigor
rigor_mexico = px.line(mexico, x="date", y="stringency_index", color='location')
rigor_brasil = px.line(brasil, x="date", y="stringency_index", color='location')
rigor_novazelandia = px.line(novazelandia, x="date", y="stringency_index", color='location')
rigor = go.Figure(data = rigor_mexico.data + rigor_brasil.data + rigor_novazelandia.data)
rigor.data[0].line.color = 'rgb(230, 5, 42)'
rigor.data[1].line.color = 'rgb(162, 227, 11)'
rigor.data[2].line.color = 'rgb(15, 32, 214)'
rigor.show()
```
